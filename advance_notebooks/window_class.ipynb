```python
from pyspark.sql import *
from pyspark.sql.window import *
from pyspark.sql.functions import *
```

```python
class Rankers:
    def __init__(self, df: DataFrame):
        self.df = df
    

    def apply_rank(self, partition_by: list, order_by: list, col_name: str = "rank") -> DataFrame:
        windows_spec = Window.partitionBy(*partition_by).orderBy(*order_by)

        df = self.df.withColumn(col_name, rank().over(windows_spec))
        return df
    

    def apply_dense_rank(self, partition_by: list, order_by: list, col_name: str = "dense_rank") -> DataFrame:
        windows_spec = Window.partitionBy(*partition_by).orderBy(*order_by)

        df = self.df.withColumn(col_name, dense_rank().over(windows_spec))
        return df
    

    def apply_row_number(self, partition_by: list, order_by: list, col_name: str = "row_number") -> DataFrame:
        windows_spec = Window.partitionBy(*partition_by).orderBy(*order_by)

        df = self.df.withColumn(col_name, row_number().over(windows_spec))
        return df
```
