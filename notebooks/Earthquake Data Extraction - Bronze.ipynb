## **Earthquake incremental Data extraction to bronze layer**

```python
# We shall request the api incremenally by giving parameters to api query
import requests
from datetime import date, timedelta
import json

# To incrementally fetch the daily data
start_date = date.today() - timedelta(days=5)
end_date = date.today() + timedelta(days=1)

url = f"https://earthquake.usgs.gov/fdsnws/event/1/query?format=geojson&starttime={start_date}&endtime={end_date}"
```

#### **_Convert the response to json payload and load it to lakehouse_**

```python
response = requests.get(url)

response_code = response.status_code
# Check the request response code
if response_code == 200:
    print("API Response Code: 200, Working fine!")
    # get the json response
    json_data = response.json()
    json_data = json_data['features']
    # print(json.dumps(json_data))

    lakehouse_path = f"/lakehouse/default/Files/Earthquake_Data/{start_date}_earthquake_data.json"

    # using 'w' for overwrite if already exist
    with open(lakehouse_path, 'w') as json_file:
        json.dump(json_data, json_file, indent=4)

    print(f"Data saved to {lakehouse_path}")
else:
    print(f"API Response Code: {response_code}, Error is fecthing the data!")
    assert 1 == 2
```
